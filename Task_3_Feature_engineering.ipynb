### Feature engineering summary

Several new features were created to improve predictive power for churn modelling.

1. Date columns were transformed into more useful duration-based features such as customer tenure, days to contract end, days to renewal, and contract length.

2. The provided feature showing the difference between December and January off-peak prices was recreated for both variable and fixed pricing.

3. Additional price-related features were engineered by aggregating pricing data at customer level, including mean, minimum, maximum, standard deviation, and range for each pricing component.

4. Price spread and volatility features were created to capture pricing structure and changes over time, which may influence churn behaviour.

5. Existing usage and forecast columns were combined into ratio and difference features to better represent consumption patterns.

6. Raw date columns were dropped after feature creation, and categorical columns were encoded to prepare the dataset for modelling.

## 3. Feature engineering

### Difference between off-peak prices in December and preceding January

Below is the code created by your colleague to calculate the feature described above. Use this code to re-create this feature and then think about ways to build on this feature to create features with a higher predictive power.

```python
price_df = pd.read_csv('price_data.csv')
price_df["price_date"] = pd.to_datetime(price_df["price_date"], format='%Y-%m-%d')
price_df.head()
```

```python
# Group off-peak prices by companies and month
monthly_price_by_id = price_df.groupby(['id', 'price_date']).agg({'price_off_peak_var': 'mean', 'price_off_peak_fix': 'mean'}).reset_index()

# Get january and december prices
jan_prices = monthly_price_by_id.groupby('id').first().reset_index()
dec_prices = monthly_price_by_id.groupby('id').last().reset_index()

# Calculate the difference
diff = pd.merge(dec_prices.rename(columns={'price_off_peak_var': 'dec_1', 'price_off_peak_fix': 'dec_2'}), jan_prices.drop(columns='price_date'), on='id')
diff['offpeak_diff_dec_january_energy'] = diff['dec_1'] - diff['price_off_peak_var']
diff['offpeak_diff_dec_january_power'] = diff['dec_2'] - diff['price_off_peak_fix']
diff = diff[['id', 'offpeak_diff_dec_january_energy','offpeak_diff_dec_january_power']]
diff.head()
```

#5. Date-based features

Raw dates are weak. Durations are more useful.

```python
# Reference date: latest date in dataset
reference_date = df[["date_activ", "date_end", "date_modif_prod", "date_renewal"]].max().max()

# Customer tenure
df["tenure_days"] = (reference_date - df["date_activ"]).dt.days

# Days until contract end
df["days_to_end"] = (df["date_end"] - reference_date).dt.days

# Days until renewal
df["days_to_renewal"] = (df["date_renewal"] - reference_date).dt.days

# Time since product modification
df["days_since_modif_prod"] = (reference_date - df["date_modif_prod"]).dt.days

# Contract length
df["contract_length_days"] = (df["date_end"] - df["date_activ"]).dt.days

# Extract month features
df["activation_month"] = df["date_activ"].dt.month
df["end_month"] = df["date_end"].dt.month
df["renewal_month"] = df["date_renewal"].dt.month
df["modif_prod_month"] = df["date_modif_prod"].dt.month

df[[
    "tenure_days", "days_to_end", "days_to_renewal",
    "days_since_modif_prod", "contract_length_days"
]].head()
```

#6. Recreate the December vs January feature

```python
# Group prices by id and date
monthly_price_by_id = price_df.groupby(["id", "price_date"]).agg({
    "price_off_peak_var": "mean",
    "price_off_peak_fix": "mean"
}).reset_index()

# Get first and last monthly prices
jan_prices = monthly_price_by_id.groupby("id").first().reset_index()
dec_prices = monthly_price_by_id.groupby("id").last().reset_index()

# Merge first and last prices
diff = pd.merge(
    dec_prices.rename(columns={
        "price_off_peak_var": "dec_price_off_peak_var",
        "price_off_peak_fix": "dec_price_off_peak_fix"
    }),
    jan_prices.rename(columns={
        "price_off_peak_var": "jan_price_off_peak_var",
        "price_off_peak_fix": "jan_price_off_peak_fix"
    }).drop(columns="price_date"),
    on="id"
)

# Price difference features
diff["offpeak_diff_dec_january_energy"] = diff["dec_price_off_peak_var"] - diff["jan_price_off_peak_var"]
diff["offpeak_diff_dec_january_power"] = diff["dec_price_off_peak_fix"] - diff["jan_price_off_peak_fix"]

diff = diff[[
    "id",
    "offpeak_diff_dec_january_energy",
    "offpeak_diff_dec_january_power"
]]

diff.head()
```

#7. Build stronger price features

This is where you stop being average.

```python
# Aggregate price statistics by customer
price_features = price_df.groupby("id").agg({
    "price_off_peak_var": ["mean", "std", "min", "max"],
    "price_peak_var": ["mean", "std", "min", "max"],
    "price_mid_peak_var": ["mean", "std", "min", "max"],
    "price_off_peak_fix": ["mean", "std", "min", "max"],
    "price_peak_fix": ["mean", "std", "min", "max"],
    "price_mid_peak_fix": ["mean", "std", "min", "max"]
}).reset_index()

# Flatten multi-level column names
price_features.columns = ["_".join(col).strip("_") for col in price_features.columns]
price_features.head()
```

#8. Create price spread and volatility features

These are often more predictive than one raw price level.

```python
# Variable price spreads
price_features["mean_peak_offpeak_var_diff"] = (
    price_features["price_peak_var_mean"] - price_features["price_off_peak_var_mean"]
)

price_features["mean_mid_offpeak_var_diff"] = (
    price_features["price_mid_peak_var_mean"] - price_features["price_off_peak_var_mean"]
)

# Fixed price spreads
price_features["mean_peak_offpeak_fix_diff"] = (
    price_features["price_peak_fix_mean"] - price_features["price_off_peak_fix_mean"]
)

price_features["mean_mid_offpeak_fix_diff"] = (
    price_features["price_mid_peak_fix_mean"] - price_features["price_off_peak_fix_mean"]
)

# Price volatility indicators
price_features["offpeak_var_range"] = (
    price_features["price_off_peak_var_max"] - price_features["price_off_peak_var_min"]
)

price_features["peak_var_range"] = (
    price_features["price_peak_var_max"] - price_features["price_peak_var_min"]
)

price_features["midpeak_var_range"] = (
    price_features["price_mid_peak_var_max"] - price_features["price_mid_peak_var_min"]
)

price_features["offpeak_fix_range"] = (
    price_features["price_off_peak_fix_max"] - price_features["price_off_peak_fix_min"]
)

price_features.head()
```

#9. Create price change ratios

Relative change is often better than absolute change.

```python
# Merge jan/dec prices into price_features
price_features = price_features.merge(diff, on="id", how="left")

# Relative price changes
price_features["offpeak_pct_change_energy"] = np.where(
    diff["offpeak_diff_dec_january_energy"].notnull(),
    price_features["offpeak_diff_dec_january_energy"] / (price_features["price_off_peak_var_mean"] + 1e-6),
    np.nan
)

price_features["offpeak_pct_change_power"] = np.where(
    diff["offpeak_diff_dec_january_power"].notnull(),
    price_features["offpeak_diff_dec_january_power"] / (price_features["price_off_peak_fix_mean"] + 1e-6),
    np.nan
)

price_features.head()
```

#10. Merge engineered price features into client data

```python
df = df.merge(price_features, on="id", how="left")
print("Shape after merging price features:", df.shape)
df.head(3)
```

#11. Create combined consumption / forecast features

We need to exploit relationships between usage columns, not just leave them raw.

```python

if "cons_12m" in df.columns and "forecast_cons_12m" in df.columns:
    df["cons_forecast_diff"] = df["forecast_cons_12m"] - df["cons_12m"]
    df["cons_forecast_ratio"] = df["forecast_cons_12m"] / (df["cons_12m"] + 1)

if "cons_gas_12m" in df.columns and "cons_last_month" in df.columns:
    df["gas_to_last_month_ratio"] = df["cons_gas_12m"] / (df["cons_last_month"] + 1)

if "forecast_cons_year" in df.columns and "forecast_discount_energy" in df.columns:
    df["discounted_forecast_value"] = df["forecast_cons_year"] * df["forecast_discount_energy"]

df.head(3)
```

#12. Convert boolean/object columns if needed

```python
# Convert has_gas to numeric if it exists
if "has_gas" in df.columns:
    df["has_gas"] = df["has_gas"].map({"t": 1, "f": 0, True: 1, False: 0})

# One-hot encode categorical columns
categorical_cols = df.select_dtypes(include=["object"]).columns.tolist()

# Don't encode id
categorical_cols = [col for col in categorical_cols if col != "id"]

df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

print("Shape after encoding:", df.shape)
```

#13. Drop raw date columns

Once durations are created, keeping raw dates is usually unnecessary for modeling.

```python
date_cols = ["date_activ", "date_end", "date_modif_prod", "date_renewal"]
existing_date_cols = [col for col in date_cols if col in df.columns]

df = df.drop(columns=existing_date_cols)
print("Dropped raw date columns:", existing_date_cols)
```
